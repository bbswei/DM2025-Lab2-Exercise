## 1. Preprocessing Steps

```python
import re
import emoji
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import RobertaTokenizer, RobertaForSequenceClassification, get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score, classification_report
from torch import nn
```

### Load Merged Data

```python
train_df = pd.read_csv("./train_processed.csv")
test_df = pd.read_csv("./test_processed.csv")
```

Classes are highly imbalanced

```python
print(train_df['emotion'].value_counts())
```

### Label Encoding

```python
label_encoder = LabelEncoder()
train_df['label'] = label_encoder.fit_transform(train_df['emotion'])
num_labels = len(label_encoder.classes_)
class_names = label_encoder.classes_
print(f"\nnumber of labels: {num_labels}\nclass names: {class_names}")
```

### Text Cleaning

```python
def clean_text(text: str) -> str:
    """
    替換 URLs, Mentions 和 Emojis。
    """
    if pd.isna(text) or not isinstance(text, str):
        return ""
    text = re.sub(r"@\w+", " <USER> ", text)
    text = re.sub(r"http\S+|www\S+", " <URL> ", text)
    text = emoji.demojize(text, delimiters=(" ", " "))
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
train_df['text'] = train_df['text'].apply(clean_text)
test_df['text'] = test_df['text'].apply(clean_text)
```

```python
train_df.head()
```

```python
test_df.head()
```

## 2. Feature Engineering Steps

### Class Imbalance Handling

```python
# Using Inversed Class Frequency to Calculate Class Weights
class_counts = train_df['label'].value_counts().sort_index().values
total_samples = class_counts.sum()
# W_i = 1 / (Count_i / Total_Samples) = Total_Samples / Count_i
class_weights = total_samples / class_counts

class_weights = class_weights / class_weights.mean()
class_weights = torch.tensor(class_weights, dtype=torch.float32)

for i, weight in enumerate(class_weights):
    print(f"  Class [{class_names[i]}] -> Weight = {weight:.4f}")
```

```python
train_texts, val_texts, train_labels, val_labels = train_test_split(
    train_df['text'],
    train_df['label'],
    test_size=0.1,
    random_state=42,
    stratify=train_df['label']
)
```

### Tokenize

```python
TOKENIZER_NAME = 'roberta-base'
tokenizer = RobertaTokenizer.from_pretrained(TOKENIZER_NAME)
```

```python
class EmotionDataset(Dataset):
    def __init__(self, texts, labels=None, tokenizer=tokenizer, max_len=128):
        self.texts = texts.to_list() if isinstance(texts, pd.Series) else texts
        self.labels = labels.to_list() if isinstance(labels, pd.Series) else labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])

        # RoBERTa Tokenization
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        item = {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten()
        }

        if self.labels is not None:
            item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item
```

```python
train_dataset = EmotionDataset(train_texts, train_labels)
val_dataset = EmotionDataset(val_texts, val_labels)
test_dataset = EmotionDataset(test_df['text'])
```

```python
BATCH_SIZE = 16
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
```

## 3. Model Implementation Steps

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
class_weights = class_weights.to(device)
```

```python
MODEL_NAME = 'roberta-base'
model = RobertaForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=num_labels)
model.to(device)
```

```python
optimizer = AdamW(model.parameters(), lr=2e-5)
EPOCHS = 5 #3 -> score: 0.6548
```

```python
# add: 計算總步數
total_steps = len(train_dataloader) * EPOCHS
```

```python
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(total_steps * 0.1), # add: 使用總步數的 10% 作為 Warmup
    num_training_steps=total_steps
)
```

```python
criterion = nn.CrossEntropyLoss(weight=class_weights)
```

```python
def train_epoch(model, data_loader, optimizer, criterion, device, scheduler):
    model.train()
    losses = []

    for batch in data_loader:
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        loss = criterion(logits, labels)
        losses.append(loss.item())

        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0) # 梯度裁剪
        optimizer.step()
        scheduler.step() # add: 在每個 batch 後更新學習率

    return np.mean(losses)
```

```python
def evaluate(model, data_loader, device):
    model.eval()

    predictions = []
    true_labels = []

    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)

            outputs = model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            # 取最大機率的索引作為預測結果
            preds = torch.argmax(logits, dim=1).cpu().numpy()
            predictions.extend(preds)

            if 'labels' in batch:
                true_labels.extend(batch['labels'].cpu().numpy())

    if true_labels:
        f1_macro = f1_score(true_labels, predictions, average='macro')
        print(f"\nMacro F1-Score: {f1_macro:.4f}")
        print(classification_report(true_labels, predictions, target_names=class_names))
        return f1_macro

    return predictions
```

```python
best_f1 = 0
for epoch in range(EPOCHS):
    print(f"\n--- Epoch {epoch + 1}/{EPOCHS} ---")
    train_loss = train_epoch(model, train_dataloader, optimizer, criterion, device, scheduler)
    print(f"Training Loss: {train_loss:.4f}")

    current_f1 = evaluate(model, val_dataloader, device)

    if current_f1 > best_f1:
        best_f1 = current_f1
        torch.save(model.state_dict(), 'best_roberta_emotion_model.pth')
        print(f"**Saved (F1: {best_f1:.4f})**")
```

```python
test_predictions_indices = evaluate(model, test_dataloader, device)
test_predictions_emotions = label_encoder.inverse_transform(test_predictions_indices)
```

```python
submission_df = pd.DataFrame({
    'id': test_df['id'],
    'emotion': test_predictions_emotions
})

SUBMISSION_FILE_PATH = 'roberta_submission2.csv'
submission_df.to_csv(SUBMISSION_FILE_PATH, index=False)
```
